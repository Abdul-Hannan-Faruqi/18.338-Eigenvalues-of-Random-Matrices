```julia
# Code 1.1

# Experiment : Generate random samples from the normal distribution
# Plot : Histogram of random samples
# Theory : The normal distribution curve

##  Experiment
t  = 100000  # trials
dx = .2       # binsize
v  = randn(t) # samples

## Plot Histogram
using Plots
pyplot()
#histogram(v, bins = -4:dx:4, normed = true, color="yellow", labels="experiment: histogram of randn")
histogram(v, nbins = 50, normed = true, color="yellow", labels="experiment: histogram of randn")

## Theory
plot!(x->exp(-x.^2/2)/√(2*π), 
      linewidth = 2.0, 
      color = "blue", 
      xlim = (-4,4),
      labels="theory: exp(-x^2/2) / √(2π)",
      title = "Bell Curve: #Trials=$t, ")
```

```julia
#semicircle.jl
#Algorithm 1.2 of Random Eigenvalues by Alan Edelman

#Experiment:  Sample random symmetric Gaussian matrices
#Plot:        Histogram of the eigenvalues
#Theory:      Semicircle as n->infinity


## Parameters
n = 1000
t = 1


## Experiment
vals = Float64[]                            # initialize result vector
for i = 1:t
    A = randn(n, n)                         # draw n by n matrix with idependent gaussians
    S = (A + A')/2                          # symmetrize matrix
    vals = append!(vals, eigvals(S))        # calculate eigenvalues and append to result vector
end





vals /= √(n/2)                              # scale eigenvalues

xlim(-3, 3) # set limits for x axis in plot
pyplot()
histogram(vals, nbins=50, normed = true, color="yellow")                 # make histogram
xvals = linspace(-2, 2)
plot!(xvals, sqrt.(4 - xvals.^2)/(2*π),color="blue",linewidth = 2.0, title="Semi-Circle Law") # plot semi circle
```
