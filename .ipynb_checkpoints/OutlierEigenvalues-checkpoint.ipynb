```python
# Load packages
using PyPlot
using Interact
```

```python
# Define convenient functions 
jordan(n) = eye(n + 1)[2:end,1:end-1]                  # Jordan block with eigenvalue 0
randnc(n) = complex(randn(n,n), randn(n,n))/sqrt(2n);  # random complex matrix
function unitarilySimilar(A)
    n = Base.LinAlg.chksquare(A)
    Q = qr(randn(n,n))[1]                              # random Haar matrix
    return eigvals(Q*jordan(n)*Q')
end;
```

```python
# Plot 1.1 is the classic numerical eigenvalue experiment
# the eigenvalues of a matrix similar to a Jordan block appear roughly on a circle of radius
# machine epsilon of 1/n.
f = figure(figsize=(6,6))
@manipulate for n = 10:10:500; withfig(f) do
        xlim(-1,1)
        ylim(-1,1)
        vals = unitarilySimilar(jordan(n))
        plot(real(vals), imag(vals), "o")
        title("n=$n r=$(round(eps()^(1/n),2))")
        tmp = plt.gca()
        tmp[:add_patch](plt.Circle((0,0), radius = eps()^(1/n), fill = false, color = "red", linewidth = 15))
    end
end

# This example has no stable outliers.  The eigenvalues interior to the cirlce are spurious -- with no permanence.
```

```python
# Plot An + σYn
f = figure()
σ = sqrt(0.5)
@manipulate for i = 10:10:500; withfig(f) do
        xlim(-2.0,3.5)
        axis(:equal)
       # M = jordan(i) + σ*randnc(i) # the Jordan matrix
        #         M = σ*randnc(i) # the zero matrix
        M = [diagm([0,0,im/3,1,2]) zeros(5, i - 5); zeros(i - 5,5) [zeros(i-6,1) eye(i-6,i-6); eye(1, i-5)]] + σ*randnc(i) #the circular matrix
        vals = eigvals(M)
        plot(real(vals), imag(vals), "o")
        tmp = plt.gca()
#         tmp[:add_patch](plt.Circle((0,0), radius = 1, fill = false, color = "red")) # the unit circle
        tmp[:add_patch](plt.Circle((0,0), radius = sqrt(1+σ^2), fill = false, color = "red")) # the outer circle in annulus
        tmp[:add_patch](plt.Circle((0,0), radius = sqrt(1-σ^2), fill = false, color = "red")) # the inner circle in annulus
        tmp[:add_patch](plt.Circle((0,1/3), radius = 0.1, fill = false, color = "green"))
        tmp[:add_patch](plt.Circle((2,0), radius = 0.1, fill = false, color = "green"))
    end
end
```

There is an intimite link between the eigenvalues in the complex plane of M and
the histogram of the singular values of M shifted by a multiple of the identity.
(I think this idea is really Girko's, but it's also the idea that underlies 
 pseudospectra.)

Simply put, z is part of the limiting spectrum of M in the complex plane if
the histogram of M-zI has non-zero height at 0.

If z is an eigenvalue of M, then 0 is a singular value of M-zI, and therefore shows up with
some positive mass when histogramming the singular values of M-zI.


```python


# Histogram of svd(An + σYn - zI)
f = figure()
σ = sqrt(0.5)
i = 1000
M = [diagm([0,0,im/3,1,2]) zeros(5, i - 5); zeros(i - 5,5) [zeros(i-6,1) eye(i-6,i-6); eye(1, i-5)]] + σ*randnc(i)
# M = σ*randnc(500)
@manipulate for realPart in -2:0.1:2, imaginaryPart = -2:0.1:2; withfig(f) do
        xlim(0,3)
        vals = svdvals(M - complex(realPart, imaginaryPart)*I)
        plt.hist(vals, bins = 50)
    end
end
```

In the case where we started with A=0 (the second commented matrix above), we obtain Girko's circular law in the 
complex plane, but also a law that we can know about the singular values.

I believe the free cumulants are exactly κ(n)=1+z^2*n

Next step is to write down the exact density for the singular values and note that the support includes 0, exactly
when z is in the unit disk --- this is how Girko's circular law was essentially proved.
